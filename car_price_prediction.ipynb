```python
#using the KNN algorithm for testing and training data
model = KNeighborsRegressor(n_neighbors=5)
```

```python
model.fit(X_train,y_train)
```

```python
train_data = model.predict(X_train)
test_data = model.predict(X_test)
#using two variables for trainining the model and testing the model
```

```python
#Train datametrics
train_mse = mean_squared_error(y_train,train_data)
train_r2 = r2_score(y_train,train_data)
print("Training data scores")
print("mean_squared_error ",train_mse)
print("r2_score ",train_r2)

```

```python
#Test datametrics
test_mse = mean_squared_error(Y_test,test_data)
test_r2 = r2_score(Y_test,test_data)
print("Test data scores")
print("mean_squared_error ",test_mse)
print("r2_score ",test_r2)
```

```python
#using inverse_transform lets transform to original data
train_data= s.inverse_transform(train_data.reshape(1,-1))
test_data = s.inverse_transform(test_data.reshape(1,-1))
plt.scatter(s.inverse_transform(y_train.values.reshape(1,-1)), train_data, label='Training Data')
plt.scatter(s.inverse_transform(Y_test.values.reshape(1,-1)), test_data, label='Testing Data')
plt.title('KNN Regression Model')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()
```
